```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 df = pd.read_csv(r"C:\Users\TIBU\Downloads\CodvedaProject\3) Sentiment dataset.csv")
df #display dataframe
```

```python
df = df.drop(columns=["Unnamed: 0.1", "Unnamed: 0", "Timestamp", "User"])
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.info()
```

```python
num_cols = ['Retweets', 'Likes', 'Year', 'Month', 'Day', 'Hour']
```

```python
df[num_cols].mean()
```

```python
df[num_cols].median()
```

```python
df[num_cols].mode().iloc[0]
```

```python
df[num_cols].std()
```

```python
df[num_cols].hist(bins=20, figsize=(12,8))
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df[num_cols])
plt.title("Boxplots of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="Hour", y="Likes", data=df)
plt.title("Scatter Plot: Hour vs Likes")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
avg_likes_platform = df.groupby("Platform")["Likes"].mean().reset_index()
avg_likes_platform = avg_likes_platform.sort_values("Likes", ascending=False)

plt.figure(figsize=(8,5))
plt.bar(avg_likes_platform["Platform"], avg_likes_platform["Likes"], color="skyblue")
plt.title("Average Likes by Platform", fontsize=14)
plt.xlabel("Platform")
plt.ylabel("Average Likes")
plt.tight_layout()
plt.savefig("barplot_avg_likes_platform.png", dpi=300)
plt.show()
```

```python
likes_trend = df.groupby("Year")["Likes"].mean().reset_index()

plt.figure(figsize=(10,6))
plt.plot(likes_trend["Year"], likes_trend["Likes"], marker="o", color="blue", label="Average Likes")
plt.title("Average Likes Trend Over Years", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Average Likes")
plt.legend()
plt.tight_layout()
plt.savefig("linechart_likes_trend.png", dpi=300)
plt.show()
```

```python
scatter_sample = df.sample(200, random_state=42)

plt.figure(figsize=(8,6))
sns.scatterplot(x="Retweets", y="Likes", hue="Platform", data=scatter_sample, alpha=0.7, palette="tab10")
plt.title("Scatter Plot: Likes vs Retweets by Platform", fontsize=14)
plt.xlabel("Retweets")
plt.ylabel("Likes")
plt.legend(title="Platform")
plt.tight_layout()
plt.savefig("scatter_likes_retweets_sample.png", dpi=300)
plt.show()
```
