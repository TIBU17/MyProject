```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\TIBU\Downloads\CodvedaProject\1) iris.csv")
df #display DataFrame
```

```python
df = df.drop_duplicates()
df #display dataframe
```

```python
df.info()
```

```python
df.mean(numeric_only=True)
```

```python
df.median(numeric_only=True)
```

```python
df.mode().iloc[0]
```

```python
df.hist(figsize=(10,8), bins=20)
plt.suptitle("Histograms of Iris Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.drop(columns="species"))
plt.title("Boxplots of Numerical Features")
plt.show()
```

```python
sns.pairplot(df, hue="species", diag_kind="hist")
plt.suptitle("Scatterplot Matrix of Iris Features", y=1.02)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="species", y="petal_length", data=df, ci=None, palette="viridis")
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.savefig("barplot_petal_length.png")  # export as image
plt.show()
```

```python
plt.figure(figsize=(8,5))
for sp in df["species"].unique():
    subset = df[df["species"] == sp]
    plt.plot(subset.index, subset["sepal_length"], label=sp)

plt.title("Sepal Length by Index")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.savefig("linechart_sepal_length.png")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=df, palette="deep")
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.savefig("scatter_sepal_vs_petal.png")
plt.show()
```
