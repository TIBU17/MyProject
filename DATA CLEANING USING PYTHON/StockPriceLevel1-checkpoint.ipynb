```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\TIBU\Downloads\CodvedaProject\2) Stock Prices Data Set.csv")
df #Display dataframe
```

```python
df = df.drop_duplicates()
df # display dataframe
```

```python
df.info()
```

```python
df.median(numeric_only=True)
```

```python
df.mean(numeric_only=True)
```

```python
df.std(numeric_only=True)
```

```python
df[['open','high','low','close','volume']].hist(figsize=(12,8), bins=50)
plt.suptitle("Distributions of Stock Prices and Volume")
plt.tight_layout()
plt.savefig("histograms_prices_volume.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['open','high','low','close','volume']])
plt.title("Boxplots of Stock Features")
plt.ylabel("Value")
plt.tight_layout()
plt.savefig("boxplots_prices_volume.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="open", y="close", data=df)
plt.title("Scatter Plot: Open vs Close")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Stock Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
avg_close = df.groupby("symbol")["close"].mean().sort_values(ascending=False).head(10)

sns.barplot(x=avg_close.index, y=avg_close.values, palette="viridis")
plt.title("Top 10 Stocks by Average Closing Price")
plt.xlabel("Stock Symbol")
plt.ylabel("Average Close Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("barplot_top10_avg_close.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(12,6))
symbols_to_plot = ["AAPL", "MSFT", "AMZN"]

for s in symbols_to_plot:
    subset = df[df['symbol'] == s]
    plt.plot(subset['date'], subset['close'], label=s)

plt.title("Closing Price Over Time (AAPL, MSFT, AMZN)")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.tight_layout()
plt.savefig("linechart_aapl_msft_amzn.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,6))
aapl = df[df['symbol'] == "AAPL"]

sns.scatterplot(x="volume", y="close", data=aapl, alpha=0.5)
plt.title("AAPL: Volume vs Closing Price")
plt.xlabel("Trading Volume")
plt.ylabel("Closing Price")
plt.tight_layout()
plt.savefig("scatter_aapl_volume_close.png", dpi=300)
plt.show()
```
